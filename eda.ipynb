```python
import pandas as pd
import os

data_dir = "data"
csv_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]

for csv_file in sorted(csv_files):
    print(f"\nInspecting: {csv_file}")
    try:
        # Read only the 'Label' column to save memory
        labels = pd.read_csv(
            os.path.join(data_dir, csv_file),
            usecols=["Label"],
            dtype={"Label": "string"},  # Avoid type inference issues
            low_memory=False
        )["Label"].unique()
        
        print("Attack types found:")
        for label in sorted(labels):
            print(f"  - {label}")
            
    except Exception as e:
        print(f"Error: {e}")
```
